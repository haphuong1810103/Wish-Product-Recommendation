```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline
```

```python
products = pd.read_csv('summer_product.csv')
products.head()
```

```python
sellers = pd.read_csv('sellers_data.csv')
sellers.head()
```

* Columns to drop: title, product_color, product_variation_size_id, origin_country, merchant_name, merchant_has_profile_picture, merchant_profile_picture, product_picture, theme, crawl_month
* Columns to be imputed: has_urgency_banner
* Columns to be categorised: urgency_text --> check MNAR between has_urgency_banner & urgency_text
* Columns to be reviewed: merchant_info_subtitle (containing merchant's number of ratings and % of positive ratings)

**Data Normalisation**


```python
products.columns
```

```python
products = products.drop(columns=['title', 'product_color', 'product_variation_size_id', 'origin_country',
                                  'merchant_title', 'merchant_has_profile_picture', 'currency_buyer',
                                  'merchant_info_subtitle', 'merchant_rating_count', 'merchant_rating',
                                  'merchant_profile_picture', 'product_url', 'product_picture', 'theme', 'crawl_month'])
products.head()
```

```python
products.info()
```

```python
# Create a variable representing the number of tags of each products
products['num_tags'] = products['tags'].str.count('\w+')
```

```python
products.nunique()
```

```python
products.isnull().sum()
```

```python
# Visualize to see how many percentage of columns are missing

plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style('darkgrid')

missing_percentage = products.isnull().mean() * 100

# Prepare the data for plotting
missing_data = pd.DataFrame({
    'Column': missing_percentage.index,
    'Missing Percentage': missing_percentage.values
})

# Create the bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Column', y='Missing Percentage', data=missing_data, palette='viridis', hue=None)
plt.xticks(rotation=90)
plt.xlabel('Column')
plt.ylabel('Missing Percentage (%)')
plt.title('Percentage of Missing Data per Column (Products Table)')

plt.show()
```

```python
# Impute 'has_urgency_banner' with 0 representing products without urgency banner
products['has_urgency_banner'] = products['has_urgency_banner'].fillna(0)
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
products_corr = products.select_dtypes(include=numerics).corr()
plt.figure(figsize=(15, 10))
sns.heatmap(products_corr, cmap='coolwarm', annot=True, annot_kws = {'size': 8})
plt.title('Correlation between Product\'s Variables')
plt.show()
```

```python
products.hist(figsize=(10,8))
plt.tight_layout()
plt.title('Distributions of Product\'s Variables')
plt.show()
```

```python
# Number of Badges vs Units Sold
plt.figure(figsize=(10, 6))
sns.boxplot(data=products, x='badges_count', y='units_sold', palette='Set2', showmeans=True, medianprops={"color": "r", "linewidth": 1})
plt.title('Distribution of Units Sold by Number of Badges')
plt.xlabel('Number of Badges')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

```python
# Units sold of each product badge, with a focus on the effect of the badge "fast shipping"
badge_local_product = products['badge_local_product'].mul(products['units_sold'])
badge_fast_shipping = products['badge_fast_shipping'].mul(products['units_sold'])
badge_product_quality = products['badge_product_quality'].mul(products['units_sold'])
badges = pd.DataFrame({'local_product': badge_local_product,
                        'fast_shipping': badge_fast_shipping,
                        'product_quality': badge_product_quality})
badges = badges.loc[~(badges==0).all(axis=1)]
```

```python
badges[badges != 0].count()
```

```python
badges.shape
```

```python
count_badges = badges[badges == 0].fillna(1).sum(axis=1)
badges_1 = badges[count_badges == 1]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=badges_1.melt(), x='variable', y='value', palette='Set2')
plt.title('Distribution of Units Sold by Badge')
plt.xlabel('Badge')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Merchant Ratings vs Units Sold
prod_merc_rating = products[['product_id', 'merchant_id', 'units_sold', 'merchant_name']]
sellers1 = sellers[['merchantid', 'rating']]
prod_merc_rating = prod_merc_rating.set_index('merchant_id')
sellers1 = sellers1.set_index('merchantid')
prod_merc_rating = pd.concat([prod_merc_rating, sellers1.loc[prod_merc_rating.index]['rating']], axis=1).reset_index()
prod_merc_rating.rename(columns={'rating':'merchant_rating'}, inplace=True)
prod_merc_rating = prod_merc_rating.sort_values(by='units_sold')
```

```python
bin_edges = np.arange(0, prod_merc_rating['merchant_rating'].max() + 0.5, 0.5)
prod_merc_rating['merc_ratings_cat'] = pd.cut(prod_merc_rating['merchant_rating'], bins=bin_edges, right=False)
prod_merc_rating['merc_ratings_cat']
prod_merc_rating[['merchant_rating', 'merc_ratings_cat']]
```

```python
merc_rate_5 = prod_merc_rating['merc_ratings_cat'][prod_merc_rating['merc_ratings_cat'].isnull()].index
prod_merc_rating['merc_ratings_cat'] = prod_merc_rating['merc_ratings_cat'].astype('str')
prod_merc_rating['merc_ratings_cat'].loc[merc_rate_5] = '5.0'
```

```python
from pandas.api.types import CategoricalDtype
cat_rating = CategoricalDtype(categories=['[2.0, 2.5)', '[2.5, 3.0)', '[3.0, 3.5)',
                                          '[3.5, 4.0)', '[4.0, 4.5)', '[4.5, 5.0)', '5.0'], ordered=True)
prod_merc_rating['merc_ratings_cat'] = prod_merc_rating['merc_ratings_cat'].astype(cat_rating)
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(data=prod_merc_rating, x='merc_ratings_cat', y='units_sold',
            palette='Set2', showmeans=True, medianprops={"color": "r", "linewidth": 1})
plt.title('Distribution of Units Sold by Merchant Rating')
plt.xlabel('Merchant Rating')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(data=prod_merc_rating, x='merc_ratings_cat', y='units_sold',
            palette='Set2', showmeans=True, medianprops={"color": "r", "linewidth": 1})
plt.title('Distribution of Units Sold by Merchant Rating (Zoomed In)')
plt.xlabel('Merchant Rating')
plt.ylabel('Units Sold')
plt.ylim(-1000, 11000)
plt.legend()
plt.show()
```
