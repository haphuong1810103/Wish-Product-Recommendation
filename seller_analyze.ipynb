```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sellers_data.csv')
```

```python
total_units_per_merchant = df.groupby('merchantid')['totalunitssold'].sum().sort_values(ascending=False)

# Get the top 5 merchants
top_5_merchants = total_units_per_merchant.head(5).reset_index()



sns.set(style = 'darkgrid')

plt.figure(figsize=(12, 6))
sns.barplot(x='merchantid', y='totalunitssold', data=top_5_merchants, palette='viridis')
plt.xlabel('Merchant ID')
plt.ylabel('Total Units Sold')
plt.title('Top 5 Merchants by Total Units Sold')
plt.xticks(rotation=45, ha='right')  
plt.show()
```
