# SHIPPING

```python
# Visualize to see how many percentage of columns are missing 

plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style('darkgrid')

missing_percentage = df.isnull().mean() * 100

# Prepare the data for plotting
missing_data = pd.DataFrame({
    'Column': missing_percentage.index,
    'Missing Percentage': missing_percentage.values
})

# Create the bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Column', y='Missing Percentage', data=missing_data, palette='viridis', hue=None)
plt.xticks(rotation=90)  
plt.xlabel('Column')
plt.ylabel('Missing Percentage (%)')
plt.title('Percentage of Missing Data per Column')

plt.show()
```

```python
# Most Common Shipping Options

shipping_option_counts = df['shipping_option_name'].value_counts().reset_index()
shipping_option_counts.columns = ['Shipping Option', 'Count']
shipping_option_counts
```

```python
sns.set_style('darkgrid')

# Calculate the average shipping price for each shipping option and sort it
avg_shipping_price = df.groupby('shipping_option_name')['shipping_option_price'].mean().reset_index()
avg_shipping_price.columns = ['Shipping Option', 'Average Shipping Price']
avg_shipping_price = avg_shipping_price.sort_values(by='Average Shipping Price', ascending=False)

# Load fonts for displaying special characters
mpl.font_manager.fontManager.addfont('Sarabun-Regular.ttf') # Thai characters
mpl.font_manager.fontManager.addfont('KhmerOS_content.ttf') # Khmer characters
mpl.rc('font', family=['Sarabun', 'Khmer OS Content', 'DejaVu Sans'])

# Create the plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=avg_shipping_price, x='Shipping Option', y='Average Shipping Price', palette='magma')

# Add annotations
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),  # 9 points vertical offset
                textcoords='offset points')

# Set plot title and labels
plt.title('Average Shipping Price for Each Shipping Option')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Shipping Option')
plt.ylabel('Average Shipping Price')

# Show the plot
plt.show()
```

# SELLER ANALYSIS

```python
# average rating

plt.figure(figsize=(10, 6))
sns.histplot(df['merchant_rating_count'], bins=30, kde=True, color='green')
plt.title('Distribution of Merchant Rating Counts')
plt.xlabel('Rating Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top sellers based on rating

plt.figure(figsize=(12, 8))
top_sellers = df[['merchant_name', 'merchant_rating']].sort_values(by='merchant_rating', ascending=False).head(10)
sns.barplot(data=top_sellers, y='merchant_name', x='merchant_rating', palette='plasma')
plt.title('Top Sellers Based on Rating')
plt.xlabel('Merchant Rating')
plt.ylabel('Merchant Name')
plt.show()
```

```python
# correlation between units sold and ad boosts
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='uses_ad_boosts', y='units_sold', palette='Set2')
plt.title('Distribution of Units Sold by Ad Boost Usage')
plt.xlabel('Uses Ad Boosts')
plt.ylabel('Units Sold')
plt.xticks([0, 1], ['No Ad Boost', 'Ad Boost'])
plt.show()
```
